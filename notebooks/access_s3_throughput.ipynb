```python
import pathlib

root = pathlib.Path("..") / "data" / "access-3340" / "8.3-7.4.3" / "msdp-c"
datafile = root / "throughput.csv"
```

```python
import pandas as pd

df = pd.read_csv(datafile)
df["dedup_rate"] = df["dedup.1"] / 100
df["cpu_usage_secs"] = df["avgelapsed"]
df["kb_transferred"] = df["totalsz"] * 1024

df
```

```python
import statsmodels.formula.api as smf

model = smf.ols(formula="cpu_usage_secs ~ kb_transferred + kb_transferred:dedup_rate - 1", data=df)
results = model.fit()
results.summary()
```
